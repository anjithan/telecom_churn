```python
params = {
    "max_depth": [1, 2, 5, 10, 20],
    "min_samples_leaf": [5, 10, 20, 50, 100],
    "max_features": [2, 3, 5, 10, 15, 20, 25],
    "n_estimators": [10, 30, 50, 100, 200]
}
```

```python

grid_search = GridSearchCV(estimator = classifier_rf,
                           param_grid = params,
                           cv=4, n_jobs=-1, verbose=1,
                           scoring="accuracy")
```
